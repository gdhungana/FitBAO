# Fitting BAO using a Gaussian Process Regression


```python
import numpy as np
import matplotlib.pyplot as plt
from regress_gp import plot_xi,read_2pt
from regress_gp import plot_gauss_model
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) #- ignore deprecation warning for scikit etc. temporarily
```

```python
#Read the correlation values for the true mock Full sample
r,xi,xivar=read_2pt('mock/treecorr_elg_0.6-0.7.fits')
# First plot just the 2PCF as is.
plot_xi(r,xi,xivar,cut=140,fit=False,color='green')

#- Fit with a GP fit upto r=140
plot_xi(r,xi,xivar,cut=140,color='green')

```

```python
#- same for the true catalog: subsample
r2,xi2,xivar2=read_2pt('Analysisa/treecorr_elg_sample_obs_0.6-0.7.fits')
# First plot just the 2PCF as is.
plot_xi(r2,xi2,xivar2,cut=140,fit=False,color='blue')

#- Fit with a GP fit upto r=140
plot_xi(r2,xi2,xivar2,cut=140,color='blue')
```

```python
#- same for the modified catalog: quickcat model
r3,xi3,xivar3=read_2pt('Analysisb/treecorr_elg_quickcat_0.6-0.7.fits')
# First plot just the 2PCF as is.
plot_xi(r3,xi3,xivar3,cut=140,fit=False,color='red')

#- Fit with a GP fit upto r=140
plot_xi(r3,xi3,xivar3,cut=140,color='red')
```

# Fitting with a Gaussian over a smooth Power law decay

```python
#- Now use the analytic fit: Full model
plot_gauss_model(r,xi,xivar,rrange=(40,140), color='green',label='True Catalog (full)')

#- read the modified catalog after observing conditions applied and do the fit there too.
plot_gauss_model(r2,xi2,xivar2,rrange=(40,140),color='blue',label='True Catalog')
#- TODO This only has the MLE fit now. Better to perform MCMC to see the full MAP and the covariances of all the model parameters
plot_gauss_model(r3,xi3,xivar3,rrange=(40,140),color='red',label='Modified Catalog')
```

Statistically not significant shift of the BAO scale on the Gaussian model. Next using GP model of True catalog as the fiducial cosmology, observe the shift with alpha parametrization on the modified GP, marginalizing over the broadband shape.
